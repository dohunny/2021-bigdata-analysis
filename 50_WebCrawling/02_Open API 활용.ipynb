### 보건복지부 - 코로나19감염현황조회서비스(XML)

```python
servicekey='wn124jhUZB7ckzgJzEE0lmTC9Y%2FCnU11mHnOVTouh1oQIFTELKYFZOXb6mx7Hqd7f1yHFLqf4%2F1BeImdF%2BYlLg%3D%3D'

url='http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?serviceKey={}&pageNo=1&numOfRows=10&startCreateDt=20210801&endCreateDt=20210816'.format(servicekey)
url
```

```python
res=requests.get(url)
res
```

```python
res.text
```

```python
temp=xmltodict.parse(res.text)
temp
```

```python
import json
result=json.loads(json.dumps(temp)) #string
result
```

```python
# 숙제
# stateDt 기준일
# decideCnt 확진자수
# clearCnt 격리해제수
# deathCnt 사망자수
# careCnt 치료중환자수
items = result['response']['body']['items']['item']
stateDt= []
decideCnt = []
clearCnt = []
deathCnt = []
careCnt = []

for item in items:
    stateDt.append(item['stateDt'])
    decideCnt.append(item['decideCnt'])
    clearCnt.append(item['clearCnt'])
    deathCnt.append(item['deathCnt'])
    careCnt.append(item['careCnt'])
```

```python
# stateDt 기준일
# decideCnt 확진자수
# clearCnt 격리해제수
# deathCnt 사망자수
# careCnt 치료중환자수
df_data={
    '기준일':stateDt,
    '확진자수':decideCnt,
    '격리해제수':clearCnt,
    '사망자수':deathCnt,
    '치료중환자수':careCnt
}
df_data
```

```python
import pandas as pd
df=pd.DataFrame(data=df_data)
df
```
