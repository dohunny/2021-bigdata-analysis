```python
data = pd.read_csv('./bigdata/area.csv',encoding='euc-kr')
data
```

```python
result['population']=0

for city in cities:
    if ' ' in city:
        sum=data[data['소재지2'].str.contains(city.split(' ')[1])]       
    else:
        sum=data[data['소재지1'].str.contains(city)]
    sum['인구수']=sum['인구수'].astype(int)
    #print(sum['카메라대수'].sum())
    result['population'][city]=sum['인구수'].sum()
result
```

```python
result['population'].sum()
```

```python
result['area']=0

for city in cities:
    if ' ' in city:
        sum=data[data['소재지2'].str.contains(city.split(' ')[1])]       
    else:
        sum=data[data['소재지1'].str.contains(city)]
    sum['면적']=sum['면적'].astype(int)
    #print(sum['카메라대수'].sum())
    result['area'][city]=sum['면적'].sum()
result
```

```python
d_concat = pd.concat([result,d2],axis=1)
d_concat
```

```python
d_concat['crime_ratio'] = d_concat['crime']/d_concat['population'] * 100
d_concat['camera_ratio'] = d_concat['camera']/d_concat['area'] * 1000000
d_concat['lamp_ratio'] = d_concat['lamp']/d_concat['area'] * 100000
d_concat
```
